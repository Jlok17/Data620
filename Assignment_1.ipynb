### **Assignment: “Hello, Graph World”:**

The purpose of this assignment is to create the nodes and edges for a basic graph such as the Krackhardt kite(as seen below). For this graph, I will be using nodes and edges that represent part of my social circle.

Kite.PNG

```python
#Packages that are used:
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#First create an empty graph
graph = nx.Graph()

#Defining the nodes
nodes = ['Josh', 'Nick', 'Andy', 'Drew', 'Max', 'Kyle', 'Brian', 'Logan', 'Reilly', 'Mohith','Vivienne'
          ,'Anne']

#Adding the nodes to the graph
graph.add_nodes_from(nodes)
```

```python

#Defining the edges
edges = [('Josh', 'Nick'), ('Josh', 'Andy'), ('Josh', 'Drew'), ('Josh', 'Kyle'),
         ('Nick', 'Drew'), ('Nick', 'Brian'), ('Nick', 'Max'), ('Andy', 'Drew'),
         ('Andy', 'Kyle'), ('Drew', 'Kyle'), ('Drew', 'Brian'), ('Drew', 'Max'),
         ('Kyle', 'Brian'), ('Brian', 'Max'), ('Brian', 'Logan'), ('Brian', 'Reilly'),
         ('Logan', 'Reilly'), ('Logan', 'Mohith'), ('Reilly', 'Mohith'), ('Josh','Vivienne')
         , ('Josh','Anne'), ('Anne','Vivienne')]

#Adding the edges to the graph
graph.add_edges_from(edges)
```

```python
#Creating the graph
nx.draw(graph, with_labels=True, node_color='lightblue', node_size=800, font_size=10, font_weight='bold', ax=plt.gca())

plt.show()
```
